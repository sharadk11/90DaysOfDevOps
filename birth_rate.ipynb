```python
# @title
import pandas as pd
from sqlalchemy import create_engine

# Replace with your uploaded filename
df = pd.read_csv('county_natality.csv')

# Show preview
display(df.head(2))

# Create SQLAlchemy engine for a persistent database file
engine = create_engine('sqlite:///my_data.db')

# Load into SQLite persistent DB
df.to_sql('birth_table', con=engine, index=False, if_exists='replace')
```

```python

from sqlalchemy import inspect

inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
query = """
SELECT
  *
FROM
  birth_table
WHERE
  County_of_Residence = 'Erie County, NY'
  OR County_of_Residence = 'Niagara County, NY'
  OR County_of_Residence = 'Chautauqua County, NY'
ORDER BY
  County_of_Residence,
  Year,
  Births DESC
"""
df = pd.read_sql_query(query, con=engine)
print(df)
```
